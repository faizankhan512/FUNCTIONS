1.What is the difference between a function and a method in Python?

-> A function is a block of reusable code that performs a specific task, defined using the def keyword. It can be called independently.

A method is a function that is associated with an object or class. It operates on an instance of the class (object) and is called using dot notation (e.g., object.method()).

So, a method is essentially a function tied to an object.

2.Explain the concept of function arguments and parameters in Python.

->In Python:

Parameters are variables defined in the function definition. They act as placeholders for the values that will be passed to the function when called.
Example: def greet(name): → name is a parameter.

Arguments are the actual values passed to the function when calling it.
Example: greet("Alice") → "Alice" is an argument.

In short, parameters are part of the function definition, and arguments are the values passed when the function is called.

5.What are iterators in Python and how do they differ from iterables?

->In Python:

Iterable:
An iterable is any object that can return an iterator. It implements the __iter__() method or has a __getitem__() method.

Examples of iterables: lists, tuples, strings, dictionaries, sets, etc.
You can loop through an iterable (e.g., using a for loop) because it can be converted into an iterator.

Iterator:
An iterator is an object that represents a stream of data and can be iterated (looped) upon. It implements two key methods:

__iter__(): Returns the iterator object itself (usually self).

__next__(): Returns the next value from the iterable. When there are no more items, it raises the StopIteration exception.
Example of creating an iterator from an iterable:

my_list = [1, 2, 3]
iterator = iter(my_list)  # Converts the list into an iterator

print(next(iterator))  # 1

print(next(iterator))  # 2

print(next(iterator))  # 3

print(next(iterator))  

# Raises StopIteration
Key Difference:
Iterable is an object that can be used to obtain an iterator (like a list or string).

Iterator is the object that actually does the iteration (using __next__()), producing the elements one by one.

In summary, an iterable is something you can loop over, while an iterator is the mechanism that does the actual looping over the iterable.

6.Explain the concept of generators in Python and how they are defined?

->A generator in Python is a special type of iterator that generates values on the fly, one at a time, instead of storing them in memory. This makes generators memory-efficient, especially for large datasets.

How Generators are Defined:

Generator Function:

Defined using the yield keyword.

The yield statement pauses the function’s execution and returns a value, saving its state to resume later.

Example:

def count_up_to(n):

    count = 1
    while count <= n:
        yield count
        count += 1

Generator Expression:

A compact version of a generator function, written in a single line using parentheses ().

Example:

squares = (x * x for x in range(5))

Key Points:

Lazy evaluation: Values are generated only when requested.

Memory-efficient: Doesn't store all values at once.

Stateful: Continues from where it left off after each yield.

Generators are ideal for iterating over large datasets without using excessive memory.





7.What are the advantages of using generators over regular functions?

->Generators offer several advantages over regular functions:

Memory Efficiency:

Generators generate values one at a time and do not store the entire sequence in memory, making them ideal for large datasets.

Lazy Evaluation:

Values are computed only when requested, improving performance for large or infinite sequences.

State Preservation:

Generators remember their state between calls, allowing for efficient iteration without recalculating values.
Cleaner Code:

Generators simplify code when dealing with large or complex iteration, avoiding the need for manual state tracking or intermediate data structures.
In short, generators are more memory-efficient, faster for large data, and reduce the need for complex code.

9.Explain the purpose and usage of the `map()` function in Python?

->The map() function in Python applies a given function to each item in an iterable (like a list or tuple) and returns an iterator that yields the results.

Syntax:

map(function, iterable)

function: A function to apply to each item.

iterable: An iterable (like a list, tuple, etc.) whose elements the function will be applied to.

Example:

numbers = [1, 2, 3, 4]

squared = map(lambda x: x ** 2, numbers)

print(list(squared))  # Output: [1, 4, 9, 16]

Purpose and Usage:

Purpose: map() is used to apply a function to each item in an iterable, transforming the items.

Usage: Commonly used for transformations or operations on every element of an iterable, often in a functional programming style.

It saves time and makes code more concise by avoiding explicit loops.

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

->1. map():

Purpose: Applies a function to each item in an iterable and returns an iterator of the results.

Usage: Transforms elements individually.

Example:

numbers = [1, 2, 3]

result = map(lambda x: x ** 2, numbers)  # Output: [1, 4, 9]

2. filter():

Purpose: Filters items in an iterable by applying a function that returns True or False, and returns an iterator with items that evaluate to True.

Usage: Selects elements based on a condition.

Example:

numbers = [1, 2, 3, 4]

result = filter(lambda x: x % 2 == 0, numbers)  # Output: [2, 4]

3. reduce():

Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

Usage: Aggregates or accumulates values into one result.

Example:

from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)  # Output: 10

Key Differences:

map() transforms each element.

filter() selects elements based on a condition.

reduce() aggregates elements into a single value.

```python
#4.Write a Python function that checks if a given number is prime or not from
# 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [n for n in range(1, 201) if is_prime(n)]
print(prime_numbers)
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up
#to a specified number of terms.?

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            fib_value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_value
        else:
            raise StopIteration

fibonacci = FibonacciIterator(10)
for number in fibonacci:
    print(number)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a
# given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_two(5):
    print(power)
```

```python
#9.Write a Python program that uses `map()` to convert a list of temperatures
# from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 10, 20, 30, 40]

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
# 10.Create a Python program that uses `filter()` to remove all the vowels from
# a given string.

def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

input_string = "Hello, World!"

filtered_string = ''.join(filter(is_not_vowel, input_string))
print(filtered_string)
```
